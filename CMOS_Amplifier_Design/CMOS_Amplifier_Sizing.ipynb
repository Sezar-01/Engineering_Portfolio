ANÁLISE DO CIRCUITO

```python
# Definir variáveis simbólicas
s, g1, g2, g3 = sp.symbols('s g1 g2 g3')
C1, C2, C3, C13, C23 = sp.symbols('C1 C2 C3 C13 C23')
gm1, gm2, gm3 = sp.symbols('gm1 gm2 gm3')
v1, v2, v3, v_in = sp.symbols('v1 v2 v3 v_in')

# Equações nodais fornecidas
eq1 = sp.Eq(v1 * g1 + v1 * s * C1 + v_in * gm1 + s * C13 * (v1 - v3), 0)
eq2 = sp.Eq(v2 * g2 + v2 * s * C2 - gm2 * v1 + s * C23 * (v2 - v3), 0)
eq3 = sp.Eq(v3 * g3 + v3 * s * C3 + gm3 * v2 + s * C13 * (v3 - v1) + s * C23 * (v3 - v2), 0)

# Resolver o sistema para v1, v2, e v3
solutions = sp.solve([eq1, eq2, eq3], (v1, v2, v3))
```

```python
# Passo 1: Aplicar a relação de magnitude C13 >> C1, C23, C3
# Substituir os termos menores por 0 para simplificar
expression_simplified = polo_1.subs({C1: 0, C23: 0, C3: 0, C2: 0})
sp.pprint(expression_simplified)

# Adicionar etapa para remover C13 * g1 * g2 do denominador
denominator = sp.denom(expression_simplified)  # Obter o denominador

# Substituir o termo específico no denominador
denominator_without_g1g2 = denominator - C13 * g1 * g2
print("\nDenominador após remover o termo C13 * g1 * g2:")
sp.pprint(denominator_without_g1g2)

# Atualizar a expressão com o novo denominador
expression_updated = g1 * g2 * g3 / denominator_without_g1g2

# Passo 2: Reorganizar o denominador para destacar C13
factored_denominator = sp.factor(denominator_without_g1g2, C13)  # Colocar C13 em evidência
print("\nDenominador com C13 em evidência:")
sp.pprint(factored_denominator)

# Passo 3: Substituir g2 e g3 para reorganização final
# Reorganizar a expressão final
polo_1 = sp.simplify(g1 * g2 * g3 / factored_denominator)
print("\nExpressão reorganizada:")
polo_1

```

```python
# Função para calcular a razão Gm/Gds para PMOS e encontrar valores próximos de 10
def calcular_gm_sobre_gds_pmos(caminho_pasta, tolerancia=0.1):
    resultados = []
    
    # Itera sobre todos os arquivos CSV na pasta
    for arquivo in os.listdir(caminho_pasta):
        # Filtra apenas os arquivos que contêm "PMOS" e "L120nm" no nome
        if arquivo.endswith('.csv') and "PMOS" in arquivo and "L120nm" in arquivo:
            # Lê o arquivo CSV
            df = pd.read_csv(os.path.join(caminho_pasta, arquivo))
            
            # Verifica se as colunas 'gm Y', 'gds Y' e 'gmoverid Y' estão presentes
            if 'gm Y' in df.columns and 'gds Y' in df.columns and 'gmoverid Y' in df.columns:
                # Calcula a razão Gm/Gds
                df['gm/gds'] = df['gm Y'] / df['gds Y']
                
                # Filtra os casos onde Gm/Gds está próximo de 10 (dentro da tolerância)
                casos_proximos_10 = df[(df['gm/gds'] >= 10 - tolerancia) & (df['gm/gds'] <= 10 + tolerancia)]
                
                # Se houver casos que atendem ao critério, adiciona ao resultado
                if not casos_proximos_10.empty:
                    resultados.append((arquivo, casos_proximos_10[['gm Y', 'gds Y', 'gm/gds', 'gmoverid Y']]))
    
    return resultados

# Função para encontrar o menor valor de gmoverid entre os casos encontrados para L120nm
def encontrar_menor_gmoverid(resultados):
    menor_valor = None
    menor_caso = None
    
    for arquivo, casos in resultados:
        # Verifica se o arquivo é de comprimento L120nm
        if "L120nm" in arquivo:
            # Encontra o caso com o menor gmoverid Y
            caso_min = casos.loc[casos['gmoverid Y'].idxmin()]
            if menor_valor is None or caso_min['gmoverid Y'] < menor_valor:
                menor_valor = caso_min['gmoverid Y']
                menor_caso = (arquivo, caso_min)
    
    return menor_caso

# Função para encontrar um caso NMOS onde gds Y seja aproximadamente 20% do valor fornecido
def encontrar_nmos_com_gds_20_porcento(caminho_pasta, gds_target, tolerancia=0.05):
    for arquivo in os.listdir(caminho_pasta):
        # Filtra apenas os arquivos que contêm "NMOS" no nome
        if arquivo.endswith('.csv') and "NMOS" in arquivo:
            # Lê o arquivo CSV
            df = pd.read_csv(os.path.join(caminho_pasta, arquivo))
            
            # Verifica se a coluna 'gds Y' está presente
            if 'gds Y' in df.columns:
                # Define o intervalo de 20% do gds_target com uma tolerância
                gds_min = gds_target * 0.20 * (1 - tolerancia)
                gds_max = gds_target * 0.20 * (1 + tolerancia)
                
                # Filtra os casos onde gds Y está aproximadamente em 20% do gds_target
                casos_proximos = df[(df['gds Y'] >= gds_min) & (df['gds Y'] <= gds_max)]
                
                if not casos_proximos.empty:
                    return arquivo, casos_proximos[['gm Y', 'gds Y', 'gmoverid Y']]
    
    return None

# Caminho da pasta onde estão os CSVs (pasta atual)
caminho_pasta = './'

# Executa a função e obtém os resultados para PMOS
resultados_pmos = calcular_gm_sobre_gds_pmos(caminho_pasta)

# Encontra o caso com o menor gmoverid entre os resultados de L120nm
menor_caso_pmos = encontrar_menor_gmoverid(resultados_pmos)

# Exibe o caso com o menor gmoverid
if menor_caso_pmos:
    print("Caso com o menor gmoverid em L120nm (PMOS):")
    print(f"Arquivo: {menor_caso_pmos[0]}")
    print(menor_caso_pmos[1])
    print("\n")
    
    # Obtém o valor de gds Y do caso selecionado
    gds_target = menor_caso_pmos[1]['gds Y']
    
    # Encontra o caso NMOS onde gds Y seja aproximadamente 20% do valor obtido
    caso_nmos = encontrar_nmos_com_gds_20_porcento(caminho_pasta, gds_target)
    
    if caso_nmos:
        print("Caso NMOS com gds aproximadamente 20% do valor obtido:")
        print(f"Arquivo: {caso_nmos[0]}")
        print(caso_nmos[1])
    else:
        print("Nenhum caso NMOS encontrado com gds aproximadamente 20% do valor alvo.")
else:
    print("Nenhum caso encontrado para L120nm com gm/gds próximo de 10.")
```

```python
# Função para calcular e plotar Gm versus Cgs (usando cgs X) em todos os arquivos PMOS
def analisar_gm_versus_cgs_pmos(caminho_pasta, limite=6.58e-8):
    resultados = []
    gm_valores = []  # Lista para armazenar todos os valores de Gm
    cgs_valores = [] # Lista para armazenar todos os valores de Cgs (cgs X)
    gm_cgs_valores = []  # Lista para armazenar todos os valores de Gm/Cgs para o gráfico adicional

    # Itera sobre todos os arquivos CSV na pasta
    for arquivo in os.listdir(caminho_pasta):
        # Filtra apenas os arquivos que contêm "PMOS" no nome
        if arquivo.endswith('.csv') and "PMOS" in arquivo:
            print(f"\nLendo o arquivo: {arquivo}")  # Confirma leitura do arquivo
            # Lê o arquivo CSV
            df = pd.read_csv(os.path.join(caminho_pasta, arquivo))
            
            # Verifica se as colunas 'gm Y' e 'cgs X' estão presentes
            if 'gm Y' in df.columns and 'cgs Y' in df.columns:
                # Adiciona os valores de Gm e Cgs (cgs X) para o gráfico
                gm_valores.extend(df['gm Y'].values)
                cgs_valores.extend(abs(df['cgs Y'].values))
                
                # Calcula a razão Gm/Cgs e adiciona à lista para o gráfico adicional
                df['gm/cgs'] = df['gm Y'] / abs(df['cgs Y'])
                gm_cgs_valores.extend(df['gm/cgs'].values)
                
                # Exibe Gm, Cgs e Gm/Cgs para cada linha
                for _, linha in df.iterrows():
                    gm = linha['gm Y']
                    cgs = linha['cgs Y']
                    gm_cgs = linha['gm/cgs']
                    print(f"Gm: {gm}, Cgs: {cgs}, Gm/Cgs: {gm_cgs}")
                
                # Filtra os casos onde Gm/Cgs >= limite
                casos_satisfazem = df[df['gm/cgs'] >= limite]
                
                # Adiciona os casos que atendem ao critério
                if not casos_satisfazem.empty:
                    resultados.append((arquivo, casos_satisfazem[['gm Y', 'cgs Y', 'gm/cgs']]))
            else:
                print(f"Arquivo {arquivo} não contém as colunas necessárias ('gm Y' e 'cgs Y').")  # Colunas ausentes

    # Plotando Gm vs Cgs (cgs X)
    plt.figure(figsize=(12, 6))

    # Gráfico Gm vs Cgs
    plt.subplot(1, 2, 1)  # Gráfico à esquerda
    plt.scatter(cgs_valores, gm_valores, color='blue', label="Gm vs Cgs (cgs Y)")
    plt.xlabel("Cgs (cgs Y)")
    plt.ylabel("Gm")
    plt.title("Relação entre Gm e Cgs (usando cgs X) em arquivos PMOS")
    plt.grid(True)
    plt.legend()

    # Gráfico Gm/Cgs vs Linha
    plt.subplot(1, 2, 2)  # Gráfico à direita
    plt.plot(gm_cgs_valores, label="Gm/Cgs (cgs Y) por linha", color='green')
    plt.axhline(y=limite, color='red', linestyle='--', label=f"Limite {limite:.2e}")
    plt.xlabel("Linha")
    plt.ylabel("Gm/Cgs (cgs X)")
    plt.title("Gm/Cgs em função de cada linha dos PMOS")
    plt.grid(True)
    plt.legend()

    plt.tight_layout()  # Ajusta o layout para que os gráficos não se sobreponham
    plt.show()

    # Verifica se algum resultado foi encontrado
    if not resultados:
        print("Nenhum caso encontrado em todos os arquivos que atenda à condição gm/cgs >= 6.58E-8.")
    else:
        print("\nCasos que atendem à condição foram encontrados.")

    return resultados

# Caminho da pasta onde estão os CSVs (pasta atual)
caminho_pasta = './'

# Executa a função e obtém os resultados
resultados = analisar_gm_versus_cgs_pmos(caminho_pasta)

# Exibe os resultados finais
for arquivo, casos in resultados:
    print(f"\nArquivo: {arquivo}")
    print(casos)
    print("\n")
```

```python
# Definições de comprimento e gmoverid escolhidos para NMOS e PMOS na condição inicial
ln_nmos = '120n'  # Comprimento do NMOS
ln_pmos = '60n'  # Comprimento do PMOS
gmoverid_nmos = 12  # gmoverid inicial para NMOS
gmoverid_pmos = 11 # gmoverid inicial para PMOS
I_pretendida = 25e-3  # Corrente pretendida de 25 mA para a operação alternada

# Extrair DataFrames para NMOS e PMOS
df_nmos = globals()[f'DF_NMOS_{ln_nmos}']
df_pmos = globals()[f'DF_PMOS_{ln_pmos}']

# Encontrar o índice mais próximo de gmoverid desejado para o NMOS e o PMOS
idx_nmos = (df_nmos['gmoverid Y'] - gmoverid_nmos).abs().idxmin()
idx_pmos = (df_pmos['gmoverid Y'] - gmoverid_pmos).abs().idxmin()

# Extrair os valores reais de Id mais próximos para calcular K_adj com a corrente pretendida
id_nmos = df_nmos['id Y'].iloc[idx_nmos]
id_pmos = abs(df_pmos['id Y'].iloc[idx_pmos])

# Calcular K_adj para NMOS e PMOS com a corrente pretendida
K_adj_nmos = I_pretendida / id_nmos
K_adj_pmos = I_pretendida / id_pmos

# Ajustar gm e gds para NMOS usando K_adj
gmn = df_nmos['gm Y'].iloc[idx_nmos] * K_adj_nmos
gdsn = df_nmos['gds Y'].iloc[idx_nmos] * K_adj_nmos

# Ajustar gm e gds para PMOS usando K_adj
gmp = df_pmos['gm Y'].iloc[idx_pmos] * K_adj_pmos
gdsp = df_pmos['gds Y'].iloc[idx_pmos] * K_adj_pmos

# Extrair VDSsat direto das tabelas para o NMOS e PMOS
vdsat_nmos = df_nmos['vdsat Y'].iloc[idx_nmos]
vdsat_pmos = abs(df_pmos['vdsat Y'].iloc[idx_pmos])

# Recalcular o novo gm/Id com a corrente pretendida para cada transistor
gmoverid_nmos_novo = gmn / I_pretendida
gmoverid_pmos_novo = gmp / I_pretendida

# Valores de Cc2 e C3
Cc2 = 0.1 * 0.0237e-12  # Capacitância compensação (em F)
cdbN = abs(df_nmos['cdd Y'].iloc[idx_nmos])*K_adj_nmos  # Capacitância drain-bulk NMOS (em F)
cdbP = abs(df_pmos['cdd Y'].iloc[idx_pmos])*K_adj_pmos  # Capacitância drain-bulk PMOS (em F)
C3 = cdbN + cdbP  # Capacitância total de saída

# Valor de cgsND
cgsND = abs(df_nmos['cgs Y'].iloc[idx_nmos]) *K_adj_nmos  # Capacitância gate-source NMOS (em F)

# Cálculo do polo
polo = (Cc2 * gmn) / ((cgsND + C3) * (Cc2 + ((cgsND * C3) / (cgsND + C3))))

# Cálculo do ganho para as duas fases com a corrente pretendida de 25 mA
ganho_fase1 = gmn / (gdsn + gdsp + 1/16)  # Incluindo a resistência de carga de 16 ohms
ganho_db_fase1 = 20 * np.log10(ganho_fase1) if ganho_fase1 > 0 else float('NaN')

ganho_fase2 = gmp / (gdsn + gdsp + 1/16)  # Incluindo a resistência de carga de 16 ohms
ganho_db_fase2 = 20 * np.log10(ganho_fase2) if ganho_fase2 > 0 else float('NaN')

# Exibir os valores
print("Valores de ganho em dB para cada fase com a corrente pretendida de 25 mA:")
print(f"Ganho (dB) na Fase 1 (apenas NMOS ligado): {ganho_db_fase1:.2f} dB")
print(f"Ganho (dB) na Fase 2 (apenas PMOS ligado): {ganho_db_fase2:.2f} dB")

print("\nValores recalculados de gm/Id para a corrente pretendida de 25 mA:")
print(f"Novo gm/Id para NMOS: {gmoverid_nmos_novo:.2f}")
print(f"Novo gm/Id para PMOS: {gmoverid_pmos_novo:.2f}")

print("\nValores de K_adj:")
print(f"K_adj para NMOS: {K_adj_nmos:.2f}")
print(f"K_adj para PMOS: {K_adj_pmos:.2f}")

print("\nValores de VDSsat extraídos para verificar a condição < 200 mV:")
print(f"VDSsat para NMOS: {vdsat_nmos:.2f} V")
print(f"VDSsat para PMOS: {vdsat_pmos:.2f} V")

print("\nValor do polo calculado:")
print(f"Polo: {polo:.2e} Hz")
```
